```python
import matplotlib.pyplot as plt

fp = open('Alice.txt','r',encoding = 'UTF-8')
line = fp.readline()
s = line.split()
my_dict = {}
for word in s:
    my_dict[str(word)] = 1


while line:
    line = fp.readline()
    s = line.split()
    for word in s:
        if word not in my_dict:
            my_dict[word] = 1
        else:
            my_dict[word] += 1

fp.close()

num = []
for key in my_dict:
    num.append(my_dict[key])


num.sort()
num.reverse()

plt.loglog(range(len(num)), num)
plt.xlabel('Rank of the word')
plt.ylabel('Number of the occurance')
```

```python
import numpy as np
from scipy import optimize

a = range(1, len(num)+1)

plt.loglog(a, num)

def test_func(x, amp, alpha):
    return amp*x**alpha

params, params_covariance = optimize.curve_fit(test_func, a, num)
print(params, params_covariance)
plt.plot(a, test_func(a, params[0], params[1]), "r*", label = 'Fitted function')
```
