```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np

im1 = mpimg.imread('IMG_4561.JPG')
plt.imshow(im1)
```

```python
im2 = mpimg.imread('IMG_4562.JPG')
im3 = mpimg.imread('IMG_4563.JPG')
im4 = mpimg.imread('IMG_4564.JPG')
im5 = mpimg.imread('IMG_4565.JPG')
```

```python
b_im1 = im1[:, :, 2]
b_im2 = im2[:, :, 2]
b_im3 = im3[:, :, 2]
b_im4 = im4[:, :, 2]
b_im5 = im5[:, :, 2]

print(b_im1)
print(b_im1.shape)
```

```python
b_im = np.stack((b_im1,b_im2))
print(b_im)
print(b_im.shape)
```

```python
print(np.mean(b_im))
b_ave = np.mean(b_im, axis=0)
plt.imshow(b_ave)
```

```python
print(np.median(b_im))
b_med = np.median(b_im, axis=0)
plt.imshow(b_med)
```

```python
im = np.stack( (im1,im2,im3,im4,im5), axis = 0)
print(im.shape)
im_ave = np.mean(im, axis = 0)
print(im_ave.shape)
print(type(im_ave))
plt.imshow(im_ave.astype(int))
```

```python
im_med = np.median(im, axis = 0)
print(im_med.shape)
print(type(im_med))
plt.imshow(im_med.astype(int))
plt.axis("off")
plt.savefig("one_5th_of_Ting_Shuang.JPG", bbox_inches='tight', pad_inches= 0)
```
